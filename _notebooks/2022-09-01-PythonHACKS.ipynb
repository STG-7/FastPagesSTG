### Defining a Dictionary Called InfoDB

```python

# Defining a dictionary named InfoDB that contains records
InfoDB = {

# These records are of String type
"FirstName": "Shaurya",
"LastName": "Goel",
"DOB": "October 29, 2007",
"Residence": "San Diego",
"Email": "shauryaggamer@gmail.com",

# These records are of lists type
"Owns Cars": ["None"],
"Owns Devices": ["I-phone", "Gaming laptop"],
"Hobbies": ["Football", "Soccer", "Video Games"]
}




```

### Iteration through InfoDB using "For Loop"

```python
# using a for loop to iterate through every key in the dictionary and print out the key and value in a certain format
for key in InfoDB:
        print(str(key) + " : " + str(InfoDB[key]))
```

### Iteration through InfoDB using "For Loop" with Indexes

```python
# x is an integer variable
# len(InfoDB) returns length of InfoDB as an integer
# range command returns a sequence of numbers, starting from 0 by default and ends (in this case) with the length of InfoDB
for x in range(len(InfoDB)):
  
  # printing out every key and value in the dictionary using x as the index
  print(list(InfoDB)[x] +" : " + str(list(InfoDB.values())[x]))
    
```

### Iterating through InfoDB using "While Loop"

```python
# defining two integer variables that are set to 0
keynumber = 0
valuenumber = 0

# setting condition that while keynumber is less than 0 follow this code:
while keynumber < len(InfoDB) :

    # setting local key and val variables to key and values form InfoDB dictionary with the indexes of keynumber nad valuenumber
    key = list(InfoDB)[keynumber]
    val = list(InfoDB.values())[valuenumber] 
    
    # printing out key and val variables with str() because there are lists in InfoDB 
    print(str(key) + ": " + str(val))
    
    # adding and reassigning keynumber and valuenumber so that the code will go on until all of the content in InfoDB is printed out
    keynumber += 1
    valuenumber += 1
```

### Iteration through InfoDB Using "Recursion"

```python
# defining a function that takes in a variables "n"
def recursive_loop(n):
    
    # setting an if statement that "if" n is less than the length of InfoDB, follow this code:
    if n < len(InfoDB):

        # printing out the key and value in the dictionary that has the index "n"
        print(list(InfoDB)[n] +" : " + str(list(InfoDB.values())[n]))
        
        # recursion occurs here, the function is called again but n is added and reassigned to a value one greater
        recursive_loop(n + 1)


recursive_loop(0)
```

### Outputing data in InfoDB in Reverse Order

```python
for key in reversed(list(InfoDB.keys())):
    print(str(key) + " : " + str(InfoDB[key]))
```

### Adding to InfoDB using User Input

```python
newkey = ""
newvalue = ""

def AddToDB():
    print("Add a key:")
    newkey = input()
    print("Add a value for key:")
    newvalue = input()
    InfoDB.update({newkey: newvalue})

AddToDB()

for key in InfoDB:
        print(str(key) + " : " + str(InfoDB[key]))
```

### Removing from InfoDB using User Input

```python
removkey = ""

def RemovToDB():
    print("Remove a key:")
    removkey = input()
    InfoDB.pop(removkey)

RemovToDB()

for key in InfoDB:
        print(str(key) + " : " + str(InfoDB[key]))
```
