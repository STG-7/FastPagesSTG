# Homework/Hacks
Consider the following code segment:
- scores1 <- [89, 78, 92, 63, 95, 88]
- scores2 <- [92, 79, 97, 63]
- scores1 <- scores2

What are the contents of *scores1* after the code segment is executed?: 

1. [89, 78, 92, 63, 95, 88]
2. [89, 78, 92, 63, 95, 88, 92, 79, 97, 63]
3. [92, 79, 97, 63, 89, 78, 92, 63, 95, 88]
4. [92, 79, 97, 63]
    
**Answer: 4. --> [92, 79, 97, 63]**

**Explanation: Data is not being appended from scores2 into scores1 therefore the outpot for scores1 is [92, 79, 97, 63]**

Consider the following code segment:
- listA <- ["Sam", "Ann"]
- listB <- ["Jamal", "Tamara"]
- listB <- listA
- listA <- listB

What are the contents of *listA* after the code segment is executed?

1. ["Sam", "Ann"]
2. ["Jamal", "Tamara"]
3. ["Sam", "Ann", "Jamal", "Tamara"]
4. ["Jamal", "Tamara", "Sam", "Ann"]

**Answer: 1. --> ["Sam", "Ann"]**

**Explanation: Data from listA (["Sam", "Ann"]) replace what is in listB. After that, the fourth line has no change on the contents of listA because the contents of listA and listB are already the same due to the previous line of code.**
