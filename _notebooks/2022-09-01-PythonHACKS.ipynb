```python

# Defining a dictionary named InfoDB that contains records
InfoDB = {
# These records are of String type
"FirstName": "Shaurya",
"LastName": "Goel",
"DOB": "October 29, 2007",
"Residence": "San Diego",
"Email": "shauryaggamer@gmail.com",
# These records are of lists type
"Owns Cars": ["Bugatti", "Tesla"],
"Owns Devices": ["I-phone", "Gaming laptop"],
"Hobbies": ["Football", "Soccer/Basketball", "Video Games"]
}




```

### Iterating through Dictionary using "For Loop"

```python
# Using a for loop to iterate through ever key in the dictionary and print out the key a semicolong and the value
for key in InfoDB:
        print(str(key) + " : " + str(InfoDB[key]))

print("loop is done")
```

### Iteration through Dictionary using For "Loop with Indexes"

```python
keynumber = 0
valuenumber = 0

for key in InfoDB:
 key = list(InfoDB)[keynumber]
 val = list(InfoDB.values())[valuenumber]
 
 if keynumber < len(InfoDB) :
    print(str(key) + ": " + str(val))
    keynumber += 1
    valuenumber += 1

else: 
    print("Loop is done")
    
```

### Iterating through Dictionary using "While Loop"

```python
keynumber = 0
valuenumber = 0
while keynumber < len(InfoDB) :
    key = list(InfoDB)[keynumber]
    val = list(InfoDB.values())[valuenumber] 
    print(str(key) + ": " + str(val))
    keynumber += 1
    valuenumber += 1

print("loop is done")
```
